```python
# Ignorar las l√≠neas que no tienen el n√∫mero esperado de campos
df = pd.read_csv('All_databases_INICIAL_Properties.csv', on_bad_lines='skip')

# Mostrar las primeras filas para verificar
df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Definir los colores para cada grupo
colores_personalizados = {'Nat-UV': 'salmon', 'drugbank': 'mediumseagreen', 'LANaPDB': 'mediumpurple', 'UNIIQUIM': 'darkorange', 'BIOFACQUIM': 'royalblue'}  # Define los colores seg√∫n los nombres de los grupos

column_group = 'DB'  # Reemplaza por el nombre de la columna que contiene los grupos
column_value = 'Total Molweight'  # Reemplaza por el nombre de la columna con los datos para el gr√°fico

# Graficar un gr√°fico de viol√≠n con seaborn y colores personalizados
plt.figure(figsize=(6, 7))
sns.violinplot(x=column_group, y=column_value, data=df, palette=colores_personalizados)
plt.show()

# Agrupar el DataFrame por la columna 'grupo' y obtener las estad√≠sticas descriptivas
estadisticas = df.groupby(column_group)[column_value].describe()

# Mostrar las estad√≠sticas
print(estadisticas)
```
